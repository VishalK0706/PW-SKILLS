Q1. What is hypothesis testing in statisticS?
* Hypothesis testing is a statistical method used to determine whether there is enough evidence in a sample to draw conclusions about a population. It involves making an initial assumption (the null hypothesis), collecting data, and then deciding whether the data provides sufficient evidence to reject that assumption in favor of an alternative hypothesis.




Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
* The null hypothesis (H₀) states there is no effect or relationship between variables. The alternative hypothesis (H₁ or Ha) states there is an effect or relationship. The null is what researchers try to disprove; the alternative is what they try to prove

Q3.What is the significance level in hypothesis testing, and why is it important?
* The significance level (α) in hypothesis testing is the probability of rejecting the null hypothesis when it is actually true (Type I error). It is usually set before the test, commonly at 0.05, meaning there is a 5% risk of a false positive. The significance level is important because it helps control the likelihood of making incorrect conclusions from data and determines whether results are statistically significant.

Q4. What does a P-value represent in hypothesis testing?
* A p-value represents the probability of obtaining results at least as extreme as those observed, assuming the null hypothesis is true. A lower p-value indicates stronger evidence against the null hypothesis.

Q5.How do you interpret the P-value in hypothesis testing?
>  A p-value indicates how likely your observed data would occur if the null hypothesis were true.

* If the p-value is less than or equal to the significance level (commonly 0.05), reject the null hypothesis—your results are statistically significant.

* If the p-value is greater than the significance level, do not reject the null hypothesis—your results are not statistically significant

Q6. What are Type 1 and Type 2 errors in hypothesis testing?
* Type 1 error (false positive) is rejecting the null hypothesis when it is actually true.
* Type 2 error (false negative) is failing to reject the null hypothesis when it is actually false.



Q7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
* One-tailed test:
Tests if a value is specifically greater or less than a point, focusing on one direction. It’s like betting on a single side of a coin.

* Two-tailed test:
Tests if a value is simply different (higher or lower) from a point, checking both directions, like watching both sides of a coin toss.


Q8. What is the Z-test, and when is it used in hypothesis testing?
* A Z-test is a statistical test used in hypothesis testing to determine if there is a significant difference between sample and population means, or between the means of two populations, when the population standard deviation is known and the sample size is large (typically over 30). It is appropriate when the data are approximately normally distributed and the population variance is known.

Q10. What is the T-distribution, and when should it be used instead of the normal distribution?
* The t-distribution is a probability distribution similar to the normal distribution but has heavier tails, meaning it gives more probability to extreme values. It is used instead of the normal distribution when the sample size is small (typically n < 30) or when the population standard deviation is unknown. As the sample size increases, the t-distribution approaches the normal distribution.

Q11. What is the difference between a Z-test and a T-test?
> In short:
* Use a Z-test for large samples with known population standard deviation.
* Use a T-test for small samples or when the population standard deviation is unknown

Q12.What is the T-test, and how is it used in hypothesis testing?
* A t-test is a statistical test used to compare the means of two groups, especially when sample sizes are small and the population standard deviation is unknown. It helps determine if the difference between group means is statistically significant in hypothesis testing.

Q13.What is the relationship between Z-test and T-test in hypothesis testing?
* The Z-test is used for large samples with known population variance, while the T-test is for small samples with unknown variance.
* Both test mean differences, but the T-test accounts for extra uncertainty in small samples.


Q14.What is a confidence interval, and how is it used to interpret statistical result?
* A confidence interval is a range of values, calculated from sample data, that likely contains the true population parameter with a certain confidence level (e.g., 95%). It shows the estimate's precision and helps interpret the reliability of statistical results.

Q15. What is the margin of error, and how does it affect the confidence interval?
* The margin of error is a statistic that shows how much sample results can differ from the true population value, reflecting the amount of random sampling error. It is expressed as a plus-or-minus value and forms the "radius" of the confidence interval—wider margins mean less precise estimates. A larger margin of error results in a wider confidence interval, indicating more uncertainty about the estimate, while a smaller margin of error creates a narrower confidence interval, showing greater precision.

Q16. How is Bayes' Theorem used in statistics, and what is its significance?
* Bayes' Theorem is a mathematical formula used in statistics to calculate the probability of an event based on prior knowledge or evidence about related events. It is central to Bayesian inference, where it updates the probability of a hypothesis as new data becomes available. In practice, Bayes' Theorem helps revise predictions or beliefs by combining prior probabilities (what was believed before new evidence) with the likelihood of current observations, resulting in a posterior probability (updated belief after considering the evidence).

* Its significance lies in its ability to incorporate both existing knowledge and new information, making it widely used in fields like medicine (interpreting test results), finance (risk assessment), machine learning (classification), and scientific research. Bayes' Theorem provides a systematic way to update probabilities and make better-informed decisions as new evidence emerges.

Q17.What is the Chi-square distribution, and when is it used?
* The chi-square distribution is a continuous probability distribution formed by summing the squares of independent standard normal variables. It is mainly used in hypothesis testing, particularly for categorical data. Common applications include the chi-square goodness of fit test (to check if observed data matches expected frequencies) and the chi-square test of independence (to determine if two categorical variables are related). The distribution’s shape depends on its degrees of freedom, and it is not typically used to model real-world measurements, but rather to assess statistical significance in categorical data analysis.

Q18. What is the Chi-square goodness of fit test, and how is it applied?
* The Chi-square goodness of fit test checks if observed categorical data matches expected frequencies from a specific distribution. It helps determine whether sample data fits a theoretical model by comparing observed and expected counts to see if differences are due to chance or are statistically significant

Q19.  What is the F-distribution, and when is it used in hypothesis testing?
> The F-distribution is a continuous, right-skewed probability distribution used primarily in hypothesis testing to compare the variances of two populations or to test if multiple group means are equal, such as in analysis of variance (ANOVA). It arises as the ratio of two independent chi-square variables divided by their respective degrees of freedom. The F-distribution is most commonly used:

* To test if two population variances are equal.

* In ANOVA, to determine if there are significant differences among group means.

* In regression analysis, to assess the overall significance of a model.

> Its shape depends on two parameters: the degrees of freedom for the numerator and denominator, which are related to sample sizes or number of groups compared

Q20.What is an ANOVA test, and what are its assumptions?
* ANOVA (Analysis of Variance) is a statistical test used to determine whether there are significant differences between the means of three or more groups by comparing the variance between groups to the variance within groups. It uses the F-test to assess if observed differences are likely due to chance or reflect real group distinctions.

> Assumptions of ANOVA:

* The data in each group are normally distributed.

* The variances across groups are equal (homogeneity of variance).

* Observations are independent of each other.

> If these assumptions are met, ANOVA provides a reliable way to test for differences among group means.

Q21.What are the different types of ANOVA tests?
> The main types of ANOVA tests are:

* One-Way ANOVA: Compares means across three or more groups based on one independent variable.

* Two-Way ANOVA: Examines the effect of two independent variables and their interaction on the dependent variable.

* Factorial ANOVA: An extension of two-way ANOVA, involving multiple factors with different levels to analyze complex interactions.

* Repeated Measures ANOVA: Used when the same subjects are measured multiple times under different conditions or over time.

* Mixed-Model ANOVA: Combines between-subjects and within-subjects factors, often used in experimental designs with repeated measures and grouping.

* ANCOVA (Analysis of Covariance): Adjusts for the effects of covariates while comparing group means.

* MANOVA (Multivariate ANOVA): Tests for differences in means across multiple dependent variables.

> These variants allow researchers to analyze data structures ranging from simple group comparisons to complex experimental designs.

Q22.What is the F-test, and how does it relate to hypothesis testing?
* The F-test is a statistical test used in hypothesis testing to compare the variances of two populations or samples, determining if they are significantly different. It uses the F-statistic, which is the ratio of two sample variances, and follows the F-distribution. The F-test is also fundamental in ANOVA and regression analysis to assess model significance and variance among groups. If the F-statistic exceeds the critical value, the null hypothesis of equal variances is rejected

```python
#Q6.Write a Python program to perform an independent T-test and interpret the results?

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu0 = 0        # Mean under H0
mu1 = 1        # Mean under H1
sigma = 0.4    # Standard deviation
alpha = 0.05   # Significance level

# Compute critical value for right-tailed test
critical_value = norm.ppf(1 - alpha, loc=mu0, scale=sigma)

# x range for plotting
x = np.linspace(-4, 4, 1000)
y0 = norm.pdf(x, mu0, sigma)
y1 = norm.pdf(x, mu1, sigma)

plt.figure(figsize=(10,6))

# Plot distributions
plt.plot(x, y0, label="H0: mu=mu0", color="#17becf", linewidth=2)
plt.plot(x, y1, label="H1: mu=mu1", color="#ffbb78", linewidth=2)

# Shade Type 1 error (alpha) region under H0
x_alpha = x[x > critical_value]
plt.fill_between(x_alpha, 0, norm.pdf(x_alpha, mu0, sigma), color='red', alpha=0.3, label="Type 1 Error (alpha)")

# Shade Type 2 error (beta) region under H1
x_beta = x[x <= critical_value]
plt.fill_between(x_beta, 0, norm.pdf(x_beta, mu1, sigma), color='blue', alpha=0.2, label="Type 2 Error (beta)")

# Critical value line
plt.axvline(critical_value, color='k', linestyle='--', linewidth=2, label="Critical Value")

# Labels and legend
plt.xlabel("Sample Mean")
plt.ylabel("Probability Density")
plt.legend()
plt.tight_layout()
plt.show()

```

```python
#Q9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance?

import numpy as np
from scipy import stats

def confidence_interval_mean(data, confidence=0.95):
    """
    Calculate the confidence interval for the sample mean.

    Parameters:
        data (array-like): Sample data.
        confidence (float): Confidence level (default 0.95 for 95% CI).

    Returns:
        (lower_bound, upper_bound): Tuple with confidence interval.
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    sem = stats.sem(data)  # Standard error of the mean
    df = n - 1
    ci = stats.t.interval(confidence, df, loc=mean, scale=sem)
    return ci

# Example usage:
sample = [104, 106, 106, 107, 107, 107, 108, 108, 108, 108, 108, 109, 109, 109, 110, 110, 111, 111, 112]
ci = confidence_interval_mean(sample, 0.95)
print(f"95% confidence interval: {ci}")
```

```python
#Q10. ED Write a Python program to calculate the margin of error for a given confidence level using sample data?

import numpy as np
from scipy.stats import norm, t

def margin_of_error(data, confidence=0.95, population_std=None):
    """
    Calculate the margin of error for a sample mean at a given confidence level.

    Parameters:
        data (array-like): Sample data.
        confidence (float): Confidence level (default 0.95).
        population_std (float or None): If known, use population std (Z); else use sample std (T).

    Returns:
        float: Margin of error.
    """
    n = len(data)
    mean = np.mean(data)
    if population_std is not None:
        # Use Z-distribution
        se = population_std / np.sqrt(n)
        z_score = norm.ppf(1 - (1 - confidence)/2)
        moe = z_score * se
    else:
        # Use T-distribution
        sample_std = np.std(data, ddof=1)
        se = sample_std / np.sqrt(n)
        t_score = t.ppf(1 - (1 - confidence)/2, df=n-1)
        moe = t_score * se
    return moe

# Example usage:
sample = [104, 106, 106, 107, 107, 107, 108, 108, 108, 108, 108, 109, 109, 109, 110, 110, 111, 111, 112]
moe = margin_of_error(sample, confidence=0.95)
print(f"Margin of error (95% confidence): {moe:.3f}")
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process?

def bayes_theorem(prior, likelihood, evidence):
    """
    Calculate the posterior probability using Bayes' Theorem.

    Parameters:
        prior (float): P(H), prior probability of hypothesis H
        likelihood (float): P(E|H), probability of evidence E given H
        evidence (float): P(E), total probability of evidence E

    Returns:
        float: Posterior probability P(H|E)
    """
    posterior = (likelihood * prior) / evidence
    return posterior

# Example: Medical diagnostic test
# Disease prevalence (prior): 1%
# True positive rate (likelihood): 95%
# False positive rate: 2%
# Evidence: P(Positive) = P(Pos|Disease)*P(Disease) + P(Pos|No Disease)*P(No Disease)
prior = 0.01
likelihood = 0.95
false_positive = 0.02
evidence = likelihood * prior + false_positive * (1 - prior)

posterior = bayes_theorem(prior, likelihood, evidence)
print(f"Probability patient has disease given positive test: {posterior:.3f}")
```

```python
#Q12. Perform a Chi-square test for independence between two categorical variables in Python?

import numpy as np
from scipy.stats import chi2_contingency

# Example contingency table: Rows = Gender (Male, Female), Columns = Preference (A, B, C)
#          A   B   C
# Male   [10, 20, 30]
# Female [ 6,  9, 17]
table = np.array([
    [10, 20, 30],
    [6,  9, 17]
])

# Perform Chi-square test
stat, p, dof, expected = chi2_contingency(table)

print("Chi-square Statistic:", stat)
print("p-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected)

# Interpretation
alpha = 0.05
if p <= alpha:
    print("Dependent (reject H0): There is a significant association between the variables.")
else:
    print("Independent (fail to reject H0): No significant association between the variables.")
```

```python
#Q13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data?

import numpy as np

def expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test given an observed contingency table.

    Parameters:
        observed (2D array-like): The observed frequency table (rows: categories of variable 1, columns: categories of variable 2)

    Returns:
        expected (2D numpy array): Expected frequency table
    """
    observed = np.array(observed)
    row_totals = observed.sum(axis=1, keepdims=True)
    col_totals = observed.sum(axis=0, keepdims=True)
    grand_total = observed.sum()
    expected = row_totals @ col_totals / grand_total
    return expected

# Example observed data (2x3 table)
observed = np.array([
    [10, 10, 20],
    [20, 20, 20]
])

expected = expected_frequencies(observed)
print("Observed frequencies:\n", observed)
print("Expected frequencies:\n", expected)
```

```python
#Q14.  Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution?

import numpy as np
from scipy.stats import chisquare

# Example: Testing if a six-sided die is fair
# Observed frequencies from rolling a die 120 times
observed = np.array([22, 20, 18, 21, 19, 20])

# Expected frequencies for a fair die (uniform distribution)
expected = np.array([20, 20, 20, 20, 20, 20])

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square Statistic: {chi2_stat:.3f}")
print(f"p-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The observed data does NOT fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed data fits the expected distribution.")
```

```python
#Q16. Implement an F-test using Python to compare the variances of two random samples?

import numpy as np
from scipy.stats import f

# Simulate two random samples
np.random.seed(42)
sample1 = np.random.normal(10, 4, 30)  # mean=10, std=4, n=30
sample2 = np.random.normal(10, 2, 30)  # mean=10, std=2, n=30

# Calculate sample variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# F-test statistic (larger variance / smaller variance)
if var1 > var2:
    F = var1 / var2
    dfn, dfd = len(sample1) - 1, len(sample2) - 1
else:
    F = var2 / var1
    dfn, dfd = len(sample2) - 1, len(sample1) - 1

# p-value (two-tailed)
p_value = 2 * min(f.cdf(F, dfn, dfd), 1 - f.cdf(F, dfn, dfd))

print(f"Sample 1 variance: {var1:.3f}")
print(f"Sample 2 variance: {var2:.3f}")
print(f"F statistic: {F:.3f}")
print(f"Degrees of freedom: {dfn}, {dfd}")
print(f"p-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject H0: The variances are significantly different.")
else:
    print("Fail to reject H0: No significant difference in variances.")
```

```python
#Q17. Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results?

import numpy as np
from scipy.stats import f_oneway

# Simulate data for three groups
np.random.seed(42)
group1 = np.random.normal(50, 5, 30)
group2 = np.random.normal(52, 5, 30)
group3 = np.random.normal(55, 5, 30)

# Perform one-way ANOVA
f_stat, p_value = f_oneway(group1, group2, group3)

print("ANOVA Results:")
print(f"F-statistic: {f_stat:.3f}")
print(f"p-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Statistically significant difference between at least one pair of group means (reject H0).")
else:
    print("Result: No statistically significant difference between group means (fail to reject H0).")
```

```python
#Q19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA?

import numpy as np
from scipy.stats import shapiro, levene

def check_anova_assumptions(*groups, alpha=0.05):
    """
    Checks normality, equal variance, and independence assumptions for ANOVA.
    Parameters:
        *groups: arrays/lists of group data
        alpha: significance level for tests (default 0.05)
    Returns:
        dict with test results
    """
    results = {}

    # 1. Normality: Shapiro-Wilk test for each group
    normality = []
    for i, group in enumerate(groups):
        stat, p = shapiro(group)
        normality.append((p > alpha))
        print(f"Group {i+1} Shapiro-Wilk p-value: {p:.4f} -> {'Normal' if p > alpha else 'Not normal'}")
    results['normality'] = all(normality)

    # 2. Homogeneity of variances: Levene's test
    stat, p = levene(*groups)
    print(f"Levene's test p-value: {p:.4f} -> {'Equal variances' if p > alpha else 'Unequal variances'}")
    results['equal_variance'] = (p > alpha)

    # 3. Independence: Reminder
    print("Independence of observations must be ensured by study design or data collection process.")
    results['independence'] = 'Check study design'

    return results

# Example usage:
group1 = np.random.normal(50, 5, 30)
group2 = np.random.normal(52, 5, 30)
group3 = np.random.normal(55, 5, 30)

check_anova_assumptions(group1, group2, group3)
```

```python
#Q20.  Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results ?

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Example data: plant growth by watering frequency and sunlight exposure
df = pd.DataFrame({
    'water': np.repeat(['daily', 'weekly'], 15),
    'sun': np.tile(np.repeat(['low', 'med', 'high'], 5), 2),
    'height': [6, 6, 6, 5, 6, 5, 5, 6, 4, 5, 6, 6, 7, 8, 7,
               3, 4, 4, 4, 5, 4, 4, 4, 4, 4, 5, 6, 6, 7, 8]
})

# Two-way ANOVA
model = ols('height ~ C(water) + C(sun) + C(water):C(sun)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)

# Visualization: boxplot for each factor combination
plt.figure(figsize=(8, 6))
sns.boxplot(x='sun', y='height', hue='water', data=df)
plt.title('Plant Height by Sunlight and Watering Frequency')
plt.show()

# Optional: Interaction plot
import statsmodels.graphics.factorplots as fp
fig = fp.interaction_plot(df['sun'], df['water'], df['height'],
                          markers=['o', 's'], ms=8)
plt.title('Interaction Plot: Sunlight x Watering')
plt.ylabel('Mean Plant Height')
plt.show()
```

```python
#Q21. Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing?

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Set degrees of freedom for numerator and denominator
dfn = 10  # numerator degrees of freedom
dfd = 50  # denominator degrees of freedom

# Generate x values for the F-distribution
x = np.linspace(0, 5, 200)
y = f.pdf(x, dfn, dfd)

# Plot the F-distribution
plt.figure(figsize=(8,5))
plt.plot(x, y, label=f'F-distribution (dfn={dfn}, dfd={dfd})', color='blue', lw=2)
plt.fill_between(x, y, alpha=0.2, color='blue')
plt.title('F-distribution Probability Density Function')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means ?

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_1samp

# Simulate data: 100 samples from N(μ=5, σ=2)
np.random.seed(42)
mu, sigma, n = 5, 2, 100
data = np.random.normal(mu, sigma, n)

# Visualize the data
plt.hist(data, bins=15, alpha=0.7, color='skyblue', edgecolor='black', density=True)
plt.title('Histogram of Simulated Normal Data')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()

# Hypothesis test: Is the mean different from 5.5?
hypothesized_mean = 5.5
t_stat, p_value = ttest_1samp(data, hypothesized_mean)

print(f"Sample mean: {np.mean(data):.3f}")
print(f"t-statistic: {t_stat:.3f}")
print(f"p-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject H0: The sample mean is significantly different from the hypothesized mean.")
else:
    print("Fail to reject H0: No significant difference between the sample mean and the hypothesized mean.")
```

```python
#Q24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results?

import numpy as np
from scipy.stats import chi2

# Simulate data
np.random.seed(42)
data = np.random.normal(loc=10, scale=3, size=30)  # true std=3

# Hypothesized variance
sigma0_sq = 9  # (std=3)^2

# Sample statistics
n = len(data)
sample_var = np.var(data, ddof=1)

# Chi-square test statistic
chi2_stat = (n - 1) * sample_var / sigma0_sq

# Two-tailed p-value
p_value = 2 * min(chi2.cdf(chi2_stat, df=n-1), 1 - chi2.cdf(chi2_stat, df=n-1))

print(f"Sample variance: {sample_var:.3f}")
print(f"Chi-square statistic: {chi2_stat:.3f}")
print(f"Degrees of freedom: {n-1}")
print(f"p-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject H0: The population variance is significantly different from the hypothesized value.")
else:
    print("Fail to reject H0: No significant difference between sample and hypothesized variance.")
```

```python
#Q25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups?

import numpy as np
import statsmodels.api as sm

# Example: Compare the proportion of purchases in two cities
# Group 1: 300 purchases out of 500
# Group 2: 400 purchases out of 500

count = np.array([300, 400])  # Number of successes in each group
nobs = np.array([500, 500])   # Number of trials in each group

# Perform two-sample proportion Z-test
stat, pval = sm.stats.proportions_ztest(count, nobs)

print(f"Z-statistic: {stat:.3f}")
print(f"p-value: {pval:.4f}")

# Interpretation
alpha = 0.05
if pval < alpha:
    print("Reject the null hypothesis: Significant difference in proportions between the two groups.")
else:
    print("Fail to reject the null hypothesis: No significant difference in proportions between the two groups.")
```

```python
#Q26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results?

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Simulate two datasets
np.random.seed(42)
data1 = np.random.normal(10, 3, 50)  # mean=10, std=3, n=50
data2 = np.random.normal(10, 5, 50)  # mean=10, std=5, n=50

# Calculate sample variances
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

# F-test statistic (larger variance / smaller variance)
if var1 > var2:
    F = var1 / var2
    dfn, dfd = len(data1) - 1, len(data2) - 1
else:
    F = var2 / var1
    dfn, dfd = len(data2) - 1, len(data1) - 1

# Calculate p-value (two-tailed)
p_value = 2 * min(f.cdf(F, dfn, dfd), 1 - f.cdf(F, dfn, dfd))

print(f"Sample variance 1: {var1:.3f}")
print(f"Sample variance 2: {var2:.3f}")
print(f"F statistic: {F:.3f}")
print(f"Degrees of freedom: {dfn}, {dfd}")
print(f"p-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject H0: The variances are significantly different.")
else:
    print("Fail to reject H0: No significant difference in variances.")

# Visualization
plt.figure(figsize=(8,5))
plt.hist(data1, bins=20, alpha=0.6, label='Dataset 1', color='skyblue', density=True)
plt.hist(data2, bins=20, alpha=0.6, label='Dataset 2', color='salmon', density=True)
plt.title('Distribution Comparison of Two Datasets')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
#Q27. Perform a Chi-square test for goodness of fit with simulated data and analyze the result?

import numpy as np
from scipy.stats import chisquare

# Simulate observed data: e.g., rolling a 6-sided die 120 times
np.random.seed(42)
observed = np.random.choice([1,2,3,4,5,6], size=120, p=[0.15,0.15,0.20,0.20,0.15,0.15])
obs_counts = np.array([(observed == i).sum() for i in range(1,7)])

# Expected frequencies for a fair die (uniform distribution)
expected_counts = np.array([120/6]*6)

print("Observed counts:", obs_counts)
print("Expected counts:", expected_counts)

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=obs_counts, f_exp=expected_counts)

print(f"\nChi-square Statistic: {chi2_stat:.3f}")
print(f"p-value: {p_value:.4f}")

# Analyze the result
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis. The observed data does NOT fit the expected distribution.")
else:
    print("Result: Fail to reject the null hypothesis. The observed data fits the expected distribution.")
```
